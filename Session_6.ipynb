```python
labels=df_visa['continent'].unique()
certified,denied=[],[]
for i in labels:
    
    con1=df_visa['continent']==i
    con2=df_visa['case_status']=='Certified'
    con3=df_visa['case_status']=='Denied'
    certi_con=con1&con2
    deni_con=con1&con3
    certified.append(len(df_visa[certi_con]))
    denied.append(len(df_visa[deni_con]))
```

```python
case_labels=sorted(df_visa['case_status'].unique())
```

```python
pd.DataFrame(zip(certified,denied),
             index=labels,
             columns=case_labels)
```
